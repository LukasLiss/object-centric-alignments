```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
import math
```

```python
filename = "evaluation-results/bpi_filtered_activity_05_variant_10_net_001_gut.txt"
```

```python
print("Start loading data")

# read data
data = []

with open(filename) as f:
    lines = f.readlines()
    for line in lines:
        values = line.split(";")
        row = [values[0], int(values[1]), int(values[2]), int(values[3]), int(values[4]), float(values[5])]
        data.append(row)

print("Done loading data")
```

```python
# helping function
def cost_of_worst_alignment(applications, offers, events):
    total = 0
    total += 3 * applications
    total += 2 * offers
    total += events
    return  total
```

```python
# 0 : variant key
# 1 : # events
# 2 : # object types
# 3 : # object instances
# 4 : alignment cost
# 5 : time

inst_to_data = dict()
# grab relevant data points
for row in data:
    if((row[1] >=6 and row[1] <= 30) or True): # filter for outlier as well
        obj_inst = row[3]
        cost = row[4]
        time = row[5]
        inst_to_data.setdefault(obj_inst, dict())
        inst_to_data[obj_inst].setdefault(cost, [])
        inst_to_data[obj_inst][cost].append(time)

all_instance_counts = list(inst_to_data.keys())
all_instance_counts.sort(reverse=True)

for instance_count in all_instance_counts:
    x = []
    y_min = []
    y_max = []
    y_avg = []

    all_cost = list(inst_to_data[instance_count].keys())
    all_cost.sort()
    for cost in all_cost:
        x.append(cost)
        y_min.append(min(inst_to_data[instance_count][cost]))
        y_max.append(max(inst_to_data[instance_count][cost]))
        y_avg.append(sum(inst_to_data[instance_count][cost]) / len(inst_to_data[instance_count][cost]))

    plt.fill_between(x, y_min, y_max, alpha=.5, linewidth=0)
    plt.plot(x, y_avg, linewidth=2, label=f"{instance_count} obj inst")

plt.xlabel("cost")
plt.ylabel("time (s)")

plt.yscale("log")

plt.legend()
plt.show()
```

```python
# 0 : variant key
# 1 : # events
# 2 : # object types
# 3 : # object instances
# 4 : alignment cost
# 5 : time

inst_to_data = dict()
# grab relevant data points
for row in data:
    if((row[1] >=6 and row[1] <= 30) or True): # filter for outlier as well
        obj_inst = row[3]
        cost = 1 - (row[4] / cost_of_worst_alignment(1, obj_inst - 1, row[1]))
        time = row[5]
        inst_to_data.setdefault(obj_inst, dict())
        inst_to_data[obj_inst].setdefault(cost, [])
        inst_to_data[obj_inst][cost].append(time)

all_instance_counts = list(inst_to_data.keys())
all_instance_counts.sort(reverse=True)

max_x = None
min_x = None
for instance_count in all_instance_counts:
    x = []
    y_min = []
    y_max = []
    y_avg = []

    all_cost = list(inst_to_data[instance_count].keys())
    all_cost.sort(reverse=True)
    for cost in all_cost:
        x.append(cost)
        if max_x == None or cost > max_x:
            max_x = cost
        if min_x == None or cost < min_x:
            min_x = cost
        y_min.append(min(inst_to_data[instance_count][cost]))
        y_max.append(max(inst_to_data[instance_count][cost]))
        y_avg.append(sum(inst_to_data[instance_count][cost]) / len(inst_to_data[instance_count][cost]))

    plt.fill_between(x, y_min, y_max, alpha=.5, linewidth=0)
    plt.plot(x, y_avg, linewidth=2, label=f"{instance_count} obj inst")

plt.xlabel("fitness")
plt.ylabel("time (s)")

plt.yscale("log")
plt.xlim(max_x, min_x)

plt.legend()
plt.show()
```

```python
# 0 : variant key
# 1 : # events
# 2 : # object types
# 3 : # object instances
# 4 : alignment cost
# 5 : time

hue_to_data = dict()
# grab relevant data points
for row in data:
    if((row[1] >=6 and row[1] <= 30) or True): # filter for outlier as well
        events = row[1]
        cost = row[4]
        time = row[5]

        hue = cost
        x = events
        y = time

        hue_to_data.setdefault(hue, dict())
        hue_to_data[hue].setdefault(x, [])
        hue_to_data[hue][x].append(y)

all_hue_counts = list(hue_to_data.keys())
all_hue_counts.sort(reverse=True)

for hue in all_hue_counts:
    x = []
    y_min = []
    y_max = []
    y_avg = []

    all_x = list(hue_to_data[hue].keys())
    all_x.sort()
    for x_value in all_x:
        x.append(x_value)
        y_min.append(min(hue_to_data[hue][x_value]))
        y_max.append(max(hue_to_data[hue][x_value]))
        y_avg.append(sum(hue_to_data[hue][x_value]) / len(hue_to_data[hue][x_value]))

    # plt.fill_between(x, y_min, y_max, alpha=.5, linewidth=0)
    plt.plot(x, y_avg, linewidth=2, label=f"cost: {hue}")

plt.xlabel("# events")
plt.ylabel("time (s)")

plt.yscale("log")

plt.legend()
plt.show()
```
